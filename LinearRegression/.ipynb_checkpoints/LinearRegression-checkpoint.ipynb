# 训练算法

```python
def linear_regression(X, y, theta, alpha, iters=400):
    """"""
    m = len(X)
    n = len(theta)
    for i in range(iters):
        cost = compute_cost(X, y, theta)
        theta = compute_gradient(X, y, theta, alpha)
    return theta, cost

def compute_cost(X, y, theta):
    m = len(X)
    cost = (1 / (2 *m)) * np.sum(np.power(np.dot(X, theta.T) - y, 2))
    return cost

def compute_gradient(X, y, theta, alpha):
    m = len(X)
    theta = theta - (alpha / m) * np.dot((np.dot(X, theta.T) - y).T, X)
    return theta
```

```python
X = np.array([[1, x] for x in data.Population])
y = np.array(data.Profit)
y_size = y.size
y = y.reshape(y_size, 1)
print(X.shape, y.shape)
theta = np.zeros((1,X.shape[1]))
alpha = 0.01
iters = 1000
theta, cost = linear_regression(X, y, theta, alpha, iters)
print(theta)
compute_cost(X, y , theta)
linear_regression(X, y, theta, alpha, iters)
```
